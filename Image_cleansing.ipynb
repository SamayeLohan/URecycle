```python
import numpy as np
import cv2
from matplotlib import pyplot as plt

from PIL import Image
from PIL import ImageFilter
```

```python
img = cv2.imread('coffee_cup.jpg')
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    
dst = cv2.fastNlMeansDenoisingColored(img_rgb,None,10,10,7,21)

plt.subplot(121),plt.imshow(img_rgb)
plt.subplot(122),plt.imshow(dst)
plt.show()
```

```python
img_2 = cv2.imread('waterbottle.jpg')
img_rgb_2 = cv2.cvtColor(img_2, cv2.COLOR_BGR2RGB)
    
dst_2 = cv2.fastNlMeansDenoisingColored(img_rgb_2,None,15,15,7,21)

plt.subplot(121),plt.imshow(img_rgb_2)
plt.subplot(122),plt.imshow(dst_2)
plt.show()
```

```python
blur = cv2.GaussianBlur(img_rgb_2,(15,15),2)

plt.subplot(121),plt.imshow(img_rgb_2)
plt.subplot(122),plt.imshow(blur)
plt.show()
```
